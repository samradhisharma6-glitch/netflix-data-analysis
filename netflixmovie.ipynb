categorizing vote_average column
4 categories popular average below avg ,not popular


```python
def categorize_col(df, col, labels):
    edges = [
        df[col].describe()['min'],
        df[col].describe()['25%'],
        df[col].describe()['50%'],
        df[col].describe()['max']
    ]

    df[col] = pd.cut(
        df[col],
        edges,
        labels=labels,
        duplicates='drop'
    )

    return df
          
```

```python
labels = ['not_popular', 'below_avg', 'popular']

categorize_col(df, 'Vote_Average', labels)

df['Vote_Average'].unique()
```
